# Postprocess-Only Notebook

Use this notebook when you already have pipeline CSV outputs and only want visual artifacts (barcodes + annotated frames).

```python
from pathlib import Path
import pandas as pd

from pipeline.visualization import save_barcodes, save_annotated_frames
```

```python
# Set paths
condensed_csv = '/path/to/detections_condensed.csv'
full_csv = '/path/to/detections_full.csv'
image_dir = '/path/to/frames_dir'
output_dir = '/path/to/output_dir'
gt_csv_path = None  # optional

out = Path(output_dir)
out.mkdir(parents=True, exist_ok=True)
```

```python
# Barcode-only from existing condensed CSV
condensed_df = pd.read_csv(condensed_csv)
barcode_files = save_barcodes(condensed_df, str(out), gt_csv_path=gt_csv_path)
barcode_files
```

```python
# Annotated-frames-only from existing full CSV
full_df = pd.read_csv(full_csv)
annotated_files = save_annotated_frames(image_dir, full_df, str(out))
len(annotated_files)
```

```python
# Quick output checks
print('barcode_pred:', (out / 'visualizations' / 'barcode_pred.png').exists())
frames_det_dir = out / 'visualizations' / 'frames_det'
print('annotated_dir_exists:', frames_det_dir.exists())
if frames_det_dir.exists():
    print('annotated_count:', len(list(frames_det_dir.glob('*.png'))))
```
